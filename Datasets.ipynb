#IRIS – klasyfikacja (3 klasy, 4 cechy)

```python
from sklearn.datasets import load_iris

iris_raw = load_iris()

iris_data = iris_raw['data']
iris_target = iris_raw['target']
iris_feature_names = [
    name.replace(' ', '_')[:-5] for name in iris_raw['feature_names']
]

df_iris = pd.DataFrame(
    data=np.c_[iris_data, iris_target],
    columns=iris_feature_names + ['class']
)

df_iris['class'] = df_iris['class'].map({
    0.0: 'setosa',
    1.0: 'versicolor',
    2.0: 'virginica'
})

df_iris.head()
df_iris.info()
```

#WINE – klasyfikacja (chemia wina)

```python
from sklearn.datasets import load_wine

wine_raw = load_wine()

wine_data = wine_raw['data']
wine_target = wine_raw['target']
wine_feature_names = [
    name.replace(' ', '_') for name in wine_raw['feature_names']
]

df_wine = pd.DataFrame(
    data=np.c_[wine_data, wine_target],
    columns=wine_feature_names + ['class']
)

df_wine['class'] = df_wine['class'].map({
    0.0: 'class_0',
    1.0: 'class_1',
    2.0: 'class_2'
})

df_wine.head()
df_wine.info()
```

#BREAST CANCER – klasyfikacja binarna

```python
from sklearn.datasets import load_breast_cancer

cancer_raw = load_breast_cancer()

cancer_data = cancer_raw['data']
cancer_target = cancer_raw['target']
cancer_feature_names = [
    name.replace(' ', '_') for name in cancer_raw['feature_names']
]

df_cancer = pd.DataFrame(
    data=np.c_[cancer_data, cancer_target],
    columns=cancer_feature_names + ['class']
)

df_cancer['class'] = df_cancer['class'].map({
    0.0: 'malignant',
    1.0: 'benign'
})

df_cancer.head()

df_cancer.info()
```

#DIGITS – obrazy 8×8 (klasyfikacja)

```python
from sklearn.datasets import load_digits

digits_raw = load_digits()

digits_data = digits_raw['data']
digits_target = digits_raw['target']

digits_feature_names = [f'pixel_{i}' for i in range(digits_data.shape[1])]

df_digits = pd.DataFrame(
    data=np.c_[digits_data, digits_target],
    columns=digits_feature_names + ['class']
)

df_digits['class'] = df_digits['class'].astype(int)

df_digits.head()
df_digits.info()
```

#LINNERUD – regresja (fitness)

```python
from sklearn.datasets import load_linnerud

linnerud_raw = load_linnerud()

linnerud_X = linnerud_raw['data']      # cechy
linnerud_y = linnerud_raw['target']    # zmienne wyjściowe

df_linnerud_X = pd.DataFrame(
    linnerud_X,
    columns=[name.replace(' ', '_') for name in linnerud_raw['feature_names']]
)

df_linnerud_y = pd.DataFrame(
    linnerud_y,
    columns=[name.replace(' ', '_') for name in linnerud_raw['target_names']]
)

df_linnerud_X.head(), df_linnerud_y.head()
df_linnerud_X.info(), df_linnerud_y.info()
```

#MAKE_CLASSIFICATION – dane syntetyczne (klasyfikacja)

```python
from sklearn.datasets import make_classification

X_class, y_class = make_classification(
    n_samples=5000,
    n_features=10,
    n_informative=5,
    n_redundant=2,
    n_classes=3,
    random_state=42
)

df_make_class = pd.DataFrame(
    X_class,
    columns=[f'feature_{i}' for i in range(X_class.shape[1])]
)

df_make_class['class'] = y_class

df_make_class.head()
df_make_class.info()
df_make_class.describe()
```

#MAKE_REGRESSION – dane syntetyczne (regresja)

```python
from sklearn.datasets import make_regression

X_reg, y_reg = make_regression(
    n_samples=1000,
    n_features=8,
    noise=10.0,
    random_state=42
)

df_make_reg = pd.DataFrame(
    X_reg,
    columns=[f'feature_{i}' for i in range(X_reg.shape[1])]
)

df_make_reg['target'] = y_reg

df_make_reg.head()
df_make_reg.info()
df_make_reg.describe()
```

#MAKE_BLOBS – klastry (unsupervised)

```python
from sklearn.datasets import make_blobs

X_blobs, y_blobs = make_blobs(
    n_samples=1500,
    centers=6,
    n_features=2,
    random_state=42
)

df_blobs = pd.DataFrame(
    X_blobs,
    columns=['x1', 'x2']
)

df_blobs['cluster'] = y_blobs

df_blobs.head()
df_blobs.info()
df_blobs.describe()
```

#MAKE_CIRCLES – nieliniowa separacja

```python
from sklearn.datasets import make_circles

X_circles, y_circles = make_circles(
    n_samples=500,
    noise=0.05,
    factor=0.5,
    random_state=42
)

df_circles = pd.DataFrame(
    X_circles,
    columns=['x1', 'x2']
)

df_circles['class'] = y_circles

df_circles.head()
df_circles.info()
df_circles.describe()
```
