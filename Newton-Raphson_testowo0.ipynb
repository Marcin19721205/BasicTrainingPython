Metoda Newtona Raphsona - pierwiastek równanania

```python
import math

def solve_cubic_for_x(y, x0=0.0, tol=1e-8, max_iter=1000):
    """
    Solve y = x^3 + x^2 - 6x - 8 for x using Newton-Raphson.
    Parameters:
        y      : target value on the LHS
        x0     : initial guess
        tol    : stopping tolerance on |x_{n+1} - x_n|
        max_iter : safety cap on iterations
    Returns:
        x_approx, iterations
    """
    x = x0
    it = 0

    while True:
        f = (x**3 + x**2 - 6*x - 8) - y      # f(x) = RHS - y
        df = 3*x**2 + 2*x - 6                # f'(x)

        # Guard: if derivative is tiny, nudge x to avoid blow-up
        if abs(df) < 1e-12:
            x += 1e-3
            df = 3*x**2 + 2*x - 6

        x_new = x - f/df

        if abs(x_new - x) < tol or it >= max_iter:
            return x_new, it

        x = x_new
        it += 1

# --- Demo usage ---
if __name__ == "__main__":
    # Example 1: y = 0  (solving x^3 + x^2 - 6x - 8 = 0)
    # This cubic actually has three real roots; Newton converges
    # to different ones depending on the initial guess.
    for guess in [-3.0, -1.0, 3.0]:
        x_sol, iters = solve_cubic_for_x(y=0.0, x0=guess)
        print(f"y=0, start {guess:>5}: x ≈ {x_sol:.8f}  (iters: {iters})")

    # Example 2: arbitrary y (e.g., y = 10)
    x_sol, iters = solve_cubic_for_x(y=10.0, x0=2.0)
    print(f"\ny=10, start 2.0: x ≈ {x_sol:.8f}  (iters: {iters})")


```
